```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Question 1
figure = plt.figure(label = "my figure")
ax = figure.add_axes([0,0,1,1])
x = [i for i in range(0,205,20)]
y =[i for i in range(0,105,10)]
ax.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("title")
plt.show()
```

```python
data = pd.read_csv('../company_sales_data.csv')
```

```python
#Question 3
plt.plot(data['total_profit'])
plt.xlabel("Month number")
plt.ylabel("Total profit")
plt.title("Monthly profit")
plt.show()
```

```python
#Question 4
plt.plot(data['total_profit'], ls = 'dotted', 
         color = 'red', marker = 'o', linewidth = 3)
plt.xlabel("Month number")
plt.ylabel("Sold units in number")
plt.title("Monthly profit")
plt.legend(['profit'], loc = 'lower right')
plt.show()
```
